#### Grouped Bar Chart

```r
average_grades <- data %>%
  group_by(Internet) %>%
  summarize(avg_G3 = mean(G3))

groupedbar_internet <- ggplot(average_grades, aes(x = Internet, y = avg_G3, fill = Internet)) +
  geom_bar(stat = "identity") +
  labs(x = "Internet Access at Home", y = "Average Final Grade (G3)")
groupedbar_internet
```

```r
# Define the variable names for faceting
facet_vars <- colnames(data_factor)

# Create a list of ggplot objects for each facet (facetted box plots)
plots <- lapply(facet_vars, function(var) {
  ggplot(data, aes(x = .data[[var]], y = G3)) +
    geom_boxplot(fill = "blue", outlier.colour = "red") +
    labs(x = var, y = "Final Grade (G3)") +
    ggtitle(var)# +
    #theme_minimal()
})

# Combine the box plots using facet_wrap
library(patchwork)

plot_list <- wrap_plots(plots, ncol = 4)
plot_list
```
